```python
import pandas as pd 
import numpy as np 
import os
from mistralai import Mistral
```

```python
tasks = [
    "vocabulary en-it",
    "vocabulary it-en",
    "grammar",
    "conversation",
    "writing",
    "example sentences"
]

topics = [
    "Conversational Phrases",
    "Idioms and Proverbs",
    "Cultural Context",
    "Regional Dialects",
    "Slang and Informal Language",
    "Business Italian",
    "Travel Italian",
    "Food and Cuisine",
    "History and Geography",
    "Media and Entertainment",
    "Science and Technology",
    "Art and Literature",
    "Sports and Leisure",
    "Health and Medicine",
    "Legal and Administrative",
    "Environment and Sustainability"
]

converstation_step = [
    "conversation start",
    "conversation middle",
    "conversation end"
]
```

```python
def mistral_api_call(system_prompt, user_prompt, client, model="mistral-large-latest"):
    chat_response = client.chat.complete(
        model = model,
        temperature=0.7,
        tool_choice=None,
        messages = [
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": user_prompt,
            },
        ]
    )

    return chat_response.choices[0].message.content
```

```python
api_key = os.environ["MISTRAL_API_KEY"]
client = Mistral(api_key=api_key)

system_prompt = """ 
You are an expert in data generation for language learning. 
Your job is to create an extensive dataset for Italian language learning.

The example should be for intermediate learners ranging from the language levels A2 to B1.
Keep your example conscise and to the point. Avoid long explanations or descriptions.

Generate three short examples for each task and topic combination and not more.
It is important that the example are not too complicated and easy to follow.
If you are asked to generate a conversation start, make sure that the user is aksing for a specific task, like "Can you help me with my grammar?" or "I want to learn vocabulary today".
When a conversation middle is asked for, make sure that the user is responding to the assistant message and not starting a new conversation.
The same applies for a conversation end.

The data should be in JSON format and structured in a way that it contains a user instruction and an assistant message. 
The your output should look like this: 
{
      "role": "user",
      "content": "Here is a user message asking for a specific task or responding to an assistant message."
    },
    {
      "role": "assistant",
      "content": "Here is the assistant message for the task and topic combination."
}

The JSON should always just contain only one user and one assistant message and not more.
Return nothing execept the JSON.
"""
```

```python
for i in range(10): 
    task = np.random.choice(tasks)
    topic = np.random.choice(topics)
    step = np.random.choice(converstation_step)
    
    user_prompt = f"Generate an example for the task '{task}' on the topic '{topic} for the converstation state '{step}'."
    response = mistral_api_call(system_prompt, user_prompt, client)
    
    print(response)
```
